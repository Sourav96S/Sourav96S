Much more usefull is the ***NumPy*** *ndarray*.While the python array provides efficient storage of array based data,Numpy ndarray adds on efficient operation of array based data.

```python
import numpy as np
np.array([1,2,2,23,4,5,6])
```

Unlike python list,NumPy array are constrained in same data type. If the types of yhe items do not match,numpy will try to upcast the value of the items.

```python
np.array([3.14,7,4,6,6,12,89,58])
```

every items other than 3.14 was a integer,but numpy has upcasted them into floating point number.

If we want to fix the data type of the resulting array,we can use the dtype keword.

```python
np.array([3.14,7.25,8,69,45,10,38], dtype='float32')
```

Unlike,python list the numpy array can be of multidimensional array.
